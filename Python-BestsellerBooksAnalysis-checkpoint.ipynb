```python
##Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] =(12,8) #To adjust the plot size 
```

```python
#Read in data
## I had to add 'r' to convert normal string to raw string 
bestsellers=pd.read_csv(r"\Users\aturk\Downloads\bestsellers with categories.csv")
bestsellers.head()
```

```python
#Check for missing data in each column
for c in bestsellers.columns:
    percent_missing = np.mean(bestsellers[c].isnull())
    print('{} - {}%'.format(c, percent_missing))
```

```python
#Check for datatypes
bestsellers.dtypes
```

```python
#Change the datatype of genre variable
bestsellers['Genre']=bestsellers['Genre'].astype('category')
```

```python
#Sort the rows based on user rating
bestsellers.sort_values(by=["User Rating"], inplace=False, ascending= False)
```

```python
#Remove ... signs between rows in the output
pd.set_option("display.max_rows", None) 
```

```python
#Drop dublicate columns
bestsellers.drop_duplicates()
```

```python
#Check inconsistencies in genre column
unique_genre=bestsellers["Genre"].unique()
print(unique_genre)
```

```python
#Finding correlations between variables
bestsellers.corr('kendall')
```

```python
#Finding correlations between variables
bestsellers.corr('spearman')
```

```python
#Finding correlations between variables
bestsellers.corr()
```

```python
#According to pearson correlation matrix, there is no linear strong correlation between variables
#However still i will plot a heatmap
corr_matrix=bestsellers.corr()
sns.heatmap(corr_matrix, annot=True)
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value
bestsellers.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = bestsellers.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for books")

plt.show()
#Now we are able to see higher correlation between variables
```

```python
correlation_mat = bestsellers.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
